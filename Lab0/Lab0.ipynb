```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np
import sklearn as sl
from sklearn import linear_model

cars = pd.read_csv('data/km_year_power_price.csv')

parameters = ['km', 'year', 'powerPS']

for par in parameters:
    params = stats.linregress(cars[par], cars['avgPrice'])
    print("Correlation value:", params[2]**2)
    
    plt.scatter(cars[par], cars['avgPrice'])
    plt.plot(cars[par], params[1] + cars[par] * params[0], 'g', label = 'fit')
    plt.legend()
    plt.show()
```
